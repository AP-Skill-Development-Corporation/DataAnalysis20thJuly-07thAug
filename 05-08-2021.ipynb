## Seaborn

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.columns
```

#### Regression plot

```python
sns.lmplot(x = "total_bill",y="tip",data=df)
```

```python
sns.lmplot(x="total_bill",y="tip",hue="sex",data=df)
```

```python
sns.lmplot(x="total_bill",y="tip",hue="sex",data=df,markers=["d","o"])
```

```python
sns.lmplot(x="total_bill",y="tip",hue="sex",data=df,markers=["p","d"],palette="BuPu_r")
```

```python
help(sns.lmplot)
```

```python
df.columns
```

```python
sns.lmplot(x="total_bill",y="tip",data=df,col="sex",row="time",hue="smoker")
```

```python
sns.lmplot(x="total_bill",y="tip",data=df,col="sex",row="time",hue="smoker",
           height=7,palette="coolwarm")
```

### Distplot

```python
sns.distplot(df["total_bill"],bins=10)
```

### HeatMap

```python
hm = df.corr()
```

```python
hm
```

```python
sns.heatmap(hm)
```

```python
sns.heatmap(hm,annot=True)
```

```python
sns.heatmap(hm,annot=True,cmap="plasma",linecolor="black",linewidths=3)
```

```python
help(sns.heatmap)
```

```python
s = sns.load_dataset("iris")
s.head()
```

```python
s.info()
```

```python
s1 = s.corr()
s1
```

```python
sns.heatmap(s1,annot=True,vmin=-0.5,vmax=1)
```

### Pairplot

```python
sns.pairplot(s,hue="species")
```

#### Joint plot

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=s)
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=s,kind="hex")
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=s,kind="reg")
```

```python
sns.regplot(x="sepal_length",y="sepal_width",data=s)
```
