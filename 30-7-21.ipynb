```python
import pandas as pd
import numpy as np
data=[{"square":num**2,"cubes":num**3} for num in range(10,21)]
df1=pd.DataFrame(data,index=np.arange(10,21))
df1
```

### Combining Data

* Merging/Concat/append different dataframes into a single dataframe

```python
pd.concat([df1,df1],axis=1)   #rows -> axis=0 , columns -> axis=1
```

```python
df1.append(df1)   #row appending
```

```python
# merge

d1= pd.DataFrame({"emp":pd.Series(["syam","kamala","bhargav","vinitha"],index=[1,2,3,4]),
                 "yr":pd.Series([2000,2001,2002,2003],index=[1,2,3,4])
                })
print(d1)
d2=pd.DataFrame({"emp":pd.Series(["syam","kamala","bhargav","lokesh"],index=[1,2,3,4]),
                "dept":pd.Series(["hr","op","ma","s"],index=[1,2,3,4])
                })
print(d2)
```

```python
pd.merge(d1,d2)  #based on emp name
```

```python
pd.merge(d1,d2,how="left")    #left join
```

```python
pd.merge(d1,d2,how="right") #right join
```
