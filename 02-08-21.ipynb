### Line plot

* plt.plot()

```python
x_axis=[10,12,14,16]
y_axis=[5,6,7,8]
plt.plot(x_axis,y_axis)
```

```python
x_axis=[10,12,14,16]
y_axis=[5,6,7,8]
plt.plot(x_axis,y_axis,label='Line')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Line Graph')
plt.legend(loc='lower right')
```

```python
help(plt.legend)
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,20)
x
```

```python
plt.plot(x,x**2,color='red',label='Line1')
plt.plot(x,x**3,color='yellow',label='Line2')
plt.plot(x,x**4,color='green',label='Line3')
plt.legend()
```

```python
help(plt.plot)
```

```python
x1=np.arange(100)
x1
```

```python
plt.plot(x1,x1+10,color='r',linestyle='--')
plt.plot(x1,x1+15,color='g',linestyle=':')
plt.plot(x1,x1+20,color='m',linestyle='-.')
plt.plot(x1,x1+25,color='b',linestyle='-')
```

```python
x2=np.linspace(10,30,50)
x2
```

```python
plt.plot(x2,np.sin(x2),color='r',marker='o',markersize=10,markerfacecolor='g')
plt.plot(x2,np.cos(x2),color='b',marker='D',markersize=10,markerfacecolor='m')
```

```python
plt.plot(x2,np.sin(x2-1))
plt.plot(x2,np.sin(x2-2))
plt.plot(x2,np.sin(x2-3))
plt.plot(x2,np.sin(x2-4))
plt.plot(x2,np.sin(x2-5))
```

### Bar Chart
* plt.bar()

```python
year=['2019 Dec','2020 Jan','2020 March','2020 July']
Goldprice=[1000,3000,5000,7000]
plt.bar(year,Goldprice,color='g')
```

```python
year=['2019 Dec','2020 Jan','2020 March','2020 July']
Goldprice=[1000,3000,5000,7000]
silver=[500,300,400,1000]
plt.bar(year,Goldprice,width=0.33)
plt.bar(year,silver,width=0.33)
```

```python
x=np.arange(5)
x
```

```python
plt.bar(x,[90,82,55,67,77],width=0.33)
plt.bar(x+0.33,[23,90,92,44,30],width=0.33)
```

### Horizontal Bar Graph
* plt.barh()

```python
plt.barh(x,[90,82,55,67,77],color='m')
```

```python
x=[1,2,3,4,5]
y=[10,55,23,44,67]
plt.bar(x,y,color=['r','gray','m','g','b'])
```
