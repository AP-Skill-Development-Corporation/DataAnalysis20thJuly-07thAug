### Count Plot

```python
df.head()
```

```python
sns.countplot(x="smoker",data=df)
```

```python
sns.countplot(x="day",data=df)
```

```python
sns.set_style("whitegrid")
sns.countplot(x="size",data=df)
```

```python
# scale and context
#sns.set_style("whitegrid")
sns.set_context("poster",font_scale=2)# poster,paper,talk,notebook
plt.figure(figsize=(20,4))
sns.countplot(x="size",data=df)
```

```python
# scale and context
sns.set_style("white")
sns.set_context("paper",font_scale=1)# poster,paper,talk,notebook
plt.figure(figsize=(10,4))
sns.countplot(x="size",data=df)
```
